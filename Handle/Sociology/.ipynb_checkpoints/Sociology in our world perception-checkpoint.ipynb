## Разбор темы:

----------

1. **Когда впервые был употреблен термин "социология"?**

Впервые термин "социология" был употреблен Огюстом Контом в 1939 (1938) году в его работе "Курс позитивной философии".

------------

2. **Как переводится слово "социология"?**

С _латинского языка_ социология переводится как "societas" - общество и _греческого языка_ «logos» - слово, знание, учение. 

Получается: "Наука об обществе"

------------

3. **Когда социология сформировалась как самостоятельная наука? Или это произошло в тот же момент, как ее данные термин употребил О.Конт?**

Нет, социология как самостоятельная наука зародилась в конце 19 - начале 20 века, тогда, когда был опеределен предмет социологической науки.

------------

4. **Что явялется объектом и предметом социологической науки?**

**Объект социологии** - является общество в целом как целостная социальная система.

**Предмет социологии** - рассмотрение общества сквозь призму интересов людей как общественных существ, стремление постичь природу социальных связей между людьми, по поводу чего они бы не возникали, законов взаимного проспособления людей друг к другу, отношений, проявляющихся в любых облостях социальной жизни, возникающих как спонтанно, так и преднамеренно. Определить всю совокупность связей и отношений, которые носят называние "социальные". Изучение закономерностей проявления социального во всех сферах общества.

**Социальное** - это сфера отношений между людьми, возникающая в процессе их взаимодействия.

------------

5. **Сформулируйте из данных терминов определение социологии**

**Социология** - это наука о законах и закономерностях развития общества как социальной системы, о функционировании этой системы через такие ее составные элементы, как личности, социальные общности, институты.

_Дополнение:_ Социология рассматривает общество в целом, исследует поведение людей как представителей больших социальных групп, но для нее закрыт и даже не интересен внутренний мир отдельного человека.

Сравнение истории и социологии:

И история и социология исследуют общество, однако оно осуществляется в различных временных рамках:

- История исследует то, что происходило в обществе в более или менее отдаленном прошлом, в то время как социология исследует общество в настоящем, использую, безусловно опыт прошлого, зачастую для того, чтобы повлиять на будущее.

------------

6. **Что является структурой в социологическом знании?**

**Структура социологического знания** - представляет собой упорядоченную систему знаний об обществе, как развивающейся социальной системе.

------------

7. **На какие типы делится социология в зависимости от целей исследования и уровня получаемого знания?**

Социология в зависимости от целей исследования и уровня получаемого знания делится на:

------------

- Теоритическую (фундаментальную)
- Эмпирическую (прикладную)

------------

**Теоритическое социологическое исследование** - исследование, направленное на анализ и обощение накопленного знания об обществе

_**Цель теоретического исследования:**_ Разработка и построение теории общества.

------------

**Эмпирическое социологическое исследование** - исследование, направленное на сбор фактического материала о социальных явлениях и процессах, происходящих в обществе и его первичная обработка.

_**Цель эмпирического исследования:**_ оценка социальной ситуации в обществе, социальное прогнозирование и выработка практических рекомендаций по управлению обществом.


------------

8. **На какие типы делится социология от уровня изучаемого общества, его масштабности?**

_Социология от уровня изучаемого общества, его масштабности делится на:_

------------

- Макросоциологическая
- Микросоциологическая

------------

**Макросоциологическая теория** - изучает крупномасштабные социальные являения (общество как систему, его социальные институты и организации, социальные группы и др.).

_**Предпосылки:**_ Она исходит из того, что, лишь поняв общество в целом, можно понять личность.

_**Примеры концепции:**_ Структурный функционализм, неоэволюционализм, неомарксизм, теория конфликта и др.

------------

**Микросоциологическая теория** - изучает общество на микроуровне и сосредоточены непосредственно на сфере социального взаимодействия. Она рассматривает межличностные отношения и процессы социальной коммуникации в малых группах, в сфере социального поведения и т.д.

_**Примеры концепции:**_ Символический интеракционализм, этнометодология, теория обмена, анализ социальных сетей и д.р.

------------

### Теория среднего уровня

**Исключения из правил:**

- "Теории среднего уровня"
- Специальные отраслевые социологичсекие теории

**Теории среднего уровня** - изучает связь между теоритическим и эмпирическим уровнями исследования.

_**Предпосылки:**_ Теории среднего уровня направлены на теоритическое осмысление той или иной социальной подсистемы общества, которое основывается на эмпирических данных, полученных в ходе прикладных исследований.

_**Примеры концепции:**_ социология политики, экономики, досуга, молодежи, семьи и брака, образования, массовой коммуникации и т.д.

------------

9. _**Когда и кем было введено понятие "теории среднего уровня"?**_

Само понятие "теории среднего уровня" предложил американский социолог Роберт Кинг Мертон в 1947 году.

## Функции социологии

10. Какие бывают функции социологии?

------------

- Позновательная функция
- Прикладная функция (+ f социального контроля, прогностическая f, функция социального планирования)
- Идеалогическая функция

------------

**Позновательная функция** - цель = теоретическое осмысление происходящих в обществе явлений и процессов, анализ их потенциальных тенденций.

------------

_**Для кого полезно и какие нюансы?**_

_**Подводные камни с кототрыми помогает справиться социология:**_

Большинство людей, как правило, представляют и оцениваяют явления окружающего нас мира с точки зрения "здравого смысла", т.е. сквозь _призму собственного опыта_, а также определенных стереотипов, утвердившихся в обществе и разделяемых большинством людей.

_**Итог:**_ Знание социологического мышления состоит в том, что на его основе _формируется самостоятельный взгляд на общество и происходящие в нем процессы_, вырабаотывается умение сопоставлять явления социальной действительности, видеть социальную перспективу.

------------

**Прикладная функция** - цель = разработать практические рекомендации по управлению обществом, по его совершенствованию на основе анализа результатов эмпирического (кокретно-социологического) исследования социальных процессов и явлений, происходящих в обществе.

_**Подфункции**_ - функция социального контроля, прогностическая функция, функция социального планирования.

------------

**Идеалогическая функция** - цель = моделирование социального поведения и управление им, манипулирование общественным сознанием.

Отсюда появляется противоречие:

_**Эгоистичная сторона:**_

Несмотря на стремление ученых-социологов к объективности в изучении общества, результаты их прикладных социологических исследования могут быть использованы соответствующими социальными группами и их представителями для реализации собственных интересов. 

_**Гуманистическая сторона:**_

- Совершенствование и укрепление общества
- Своевременное выявление социальных противоречий =>
- Гармонизация общественных отношений
- Улучшение взаимопонимания между людьми
