# Разбор темы:

1. В чем суть учения О.Конта о трех стадиях развития общества?

Огюст Конт (1798 - 1857 гг.) сформулировал закон о 3-ех последовательных стадиях развития человечества, тождественных 3-ем стадиям его интеллектуальной эволюции:

---------

- Теологической (религиозной)
- Метафизической (философской)
- Позитивной (научной)

---------

_**1-ая - Теологическая стадия - Религиозная**_

Человек объясняет все явления, происходящие в природе и в обществе, исходя из религиозных представлений, оперируя понятием сверхъестественного.

_**2-ая - Метафизическая стадия - Философская**_

Человек отказывается от идеи божественного и пытается объяснить все при помощи абстрактных философских сущностей. Критически переосмыслив представления первой стадии, человеческая мысль этого этапа подготавливается к наступлению третьего этапа:

_**3-ья - Позитивная стадия - Научная**_

Создается социальная теория общества, лишенная теологичности и метафизичности и, обладающая чертами оперативности, доступности для непрофессионалов, практический действенности.

_**Особенности:**_ 

- На позитивной стадии человек получает свои знания как бы "снизу", с самого общества, с реальных явлений и фактов, наблюдая их и фиксируя постоянные связи, которые могут устанавливаться между ними.

- Наука не должна задаваться вопросом - почему происходит явление - а только ограничиваться описанием того, как оно происходит.

_**Задача (по смыслу Огюста Конта):**_ Описание, систематизацию и классификацию конкретных результатов и выводов научного познания.

---------

**В каком виде была представленна классификация наук по Огюсту Конту?**

В разработанной Огюстом Контом классификации наук основные науки представлены в следующем порядке:

---------

1. Математика
2. Физика (+ Астрономия)
3. Химия
4. Биология (физиология)
5. социология (социальная физика)

---------

_**Идея:**_ Все последующие науки проедставляются основанными на законах, открытых в предшествующих науках.


_**Объяснение (если кому-то требуется)**_

1. Положительные науки основаны на измерении, поэтому первая наука - **математика**
2. **Физика** основана на математике и основе механики
3. **Химическим телам** присущи механические законы и в них раскрывается новый естественный закон: кратного соотношения частей в молекулярных составных еденицах.
4. **Жизненным (биологическим) телам** присущи законы механические и химические и в изучении их открывается новый естественный закон: организации роста, закон развития и эволюции.
5. И, наконец, все эти законы присущи **социальным явлениям**

Место социологии - на вершине классификации составляющих наук и она призвана описать законы общественного развития.

---------

**На какие составные части делится социология по Огюсту Конту?**

Огюст Конст, в соответсвии со своими теоритическими представлениями делится социологию, на 2 составные части:

---------

1. Социальная статика
2. Социальная динамика

---------

2. _**Что изучает социальная статика и социальная динамика в социологии О. Конта?**_

---------

**Социальная статика** - изучает условия стабильного существования обществ, т.е. существующие структуры общества (государство, религию, семью и т.д.) с точки зрения их общественных функций в как бы застывшем (статическом) состоянии.

**Социальная динамика** - изучает теорию исторического социального развития, базирующуюся на вере в прогресс умственного развития человечества и признании закономерного прохождения стадий его развития.

_**Особенность (пояснение):**_ Социальная динамика исследует те же самые явления и структуры общества, что и социальная статика, но уже в состоянии их развития и социальных изменений, изучает законы развития общества и определяет направления общественного прогресса.

---------
_Наблюдение:_

- Социальная статика - объект исследования социологии
- Социальная динамика - предмет исследования социологии

## Социология Герберта Спенсера (1820-1903 гг.)

---------

3. _**В чем суть социального эволюционизма Г.Спенсера?**_

---------

Основная суть в том, что Гербер Спенсер взял позитивистскую социологию О.Конта и положил в основу своего мировоззрения принцип эволюции общества, функционирующего по аналогии с природой на базе законов биологии, и, создавшего **органическую теорию общества (органицизм).**

_**Основные идеи:**_

- Сословно-кастовое разделение (строение) общества и присущие ему различные функции - разделение функций между органами живого тела.


_**Исключения из идей:**_

- Отдельные личности обладают гораздо большей самостоятельностью, нежели биологические клетки
- + Подчеркивал свойство живой материи к **саморегуляции** =>
- Ставил под сомнение роль и значение государарственных форм как интсрументов **регуляции** общества, рассматривая их в значительной степени как инструменты насилия.

---------

_Зачем нужно государства, которое будет регулировать общество, если по моей теории оно как организм - саморегулируемо?_

---------

_**В основе его теории лежат 3 основных принципа:**_

---------

- "естественный отбор"
- "борьба за существование"
- "выживание сильнейшего"

---------

**"Закон выживания":**

**Государство НЕ вмешивается в естественно протекающие процессы в обществе** =>

_В таких условиях люди разделятся на 2 типа:_
- "Приспособленные"   - адаптируются и выживают
- "Неприспособленные" - вымирают

=> общество будет достигать все более выскоих уровней исторического развития

**Государство вмешивается в естественно протекающие процессы в обществе** =>

формируется наихудший тип общества с выживанием и процветанием слабейших, то есть людей с низшими интеллектуально-моральными качествами, и его последующей деградации.

**Комментарий:**

Идея изменчивости и "плавного" эволюционизма в социологии Герберта Спенсера позволяла применить к ней естественно-научные методы, что и составляло одну из целей позитивистского подхода к обществознанию.

_Герберт Спенсер считал, что общество имеет 2 полюса эволюции:_

---------

- Военный
- Промышленный

---------

**Военный тип эволюции** - характеризуется истреблением или порабощение побежденного победителем.

**Промышленный тип эволюции** - характеризуется промышленной конкуренцией в своей основе.

_**По его мнению (Герберта Спенсера):**_ Эволюция идет от первого типа ко второму. По мере того, как "закон выживания" наиболее приспособленного реализует себя в общественной динамике, общество естественным образом приближается к промышленному типу, характеризующемуся диффиренциацией, основанной на личной свободе.

---------

_**Итог:**_ Гербер Спенсер, также как и Огюст Конт, считал, что новая наука - социология - должна противопоставить радикальным теориям революции теорию, позволяющую проводить изменения в обществе эволюционным путем, учитывая все общественные факторы и интересы всех социальных групп.


## Социология Эмиля Дюркгейма (1858 - 1917 гг.)

**Один из девизов:**

4. _**Что такое социальный факт в социологии Э.Дюркгейма?**_
5.1. _**Что, по мнению Э. Дюркгейма, является предметом социологии?**_

---------

Начнем с того, что по его мнению в основе общества лежит некоторая **"особая реальность" ("социальная реальность")**, которая соединяет людей в нечто единое, устанавливает между ними солидарность особого рода и потому имеет свои общие и частные законы.

**Законы** - **предмет** чисто социологического изучения, ибо они раскрывают суть и характер всех общественных явлений и носят **объективный характер**, что и является высшим критерием научности.

**"Социальный фактор"** - предмет социологии, который выражется в соединении общего, "коллективного", но связанного с индивидуальным (т.к. только деятельность отдельных людей может наполнять общее его содержанием). 

_**Особенность социального фактора:**_ 
Социальный факт специфичен, так как он порожден объединенными действиями индивидов, но отличается от индивидуального сознания тем, что у него другое основание - **коллективное сознание**

---------

5.2. _**Что такое коллективное сознание?**_

---------

**Коллективное сознание** - совокупность общих у членов одного и того же общества интересов, верований, убеждений, чувств, ценностей и стремлений. Оно имеет особую, **"отдельную реальность"**.

"Отдельная реальность" - существует объективно, независимо от воли и сознания индивидов, но при этом осуществляется только в индивидах, которые притягиваютяс друг к другу благодаря общим верованиям и сходным чувствам.

**Особенности:**
- Является важнейшей предпосылкой духовного развития коллектива
- Чем больше общественное сознание как "голос общественной совести" регламентриует социальную жизнь общества, тем сильнее и крепче связь индивида с группой, тем устойчивее общество.

---------

_Эмиль Дюркгейм ввел такое новое для социологии понятие, как:_

**Аномия (паталогия общества)** - ощущение отсутствия норм, возникающее в переходные и кризисные периоды, когда старые нормы и ценности перестают действовать, а новые еще не установились. 

---------

6. _**Что такое солидарность в социологии Э. Дюркгейма? Чем характеризуется механическая и органическая солидарность?**_

---------

_Одна из центральных проблем в социологии Эмиля Дюркгейма - **"социальная солидарность"**_

**Социальная солидарность** - главная сила, созидающая общество как целое и способствует его сохранению. Она представляет собой логическое следствие общественного разделения труда (т.е. профессиональной специализации), которое вносит разнообразие в общественные отношения.

_Эмиль Дюркгейм выделил 2 типы социальной солидарности:_

---------

- Органическая солидарность
- Механическая солидарность

---------

**Механическая солидарность** - характерна для архаического (доиндустриального) общества. 

_**В основе:**_ Неразвитости общества и сходство индивидов и их общественных функций, что обеспечивает взаимозаменяемость индивидов.

_**Особенность механической солидарности:**_
Стремление полностью подчинить себе индивида через систему жесткого социального контроля.

---------

**Органическая солидарность** - характерна для современных обществ.

_**В основе:**_ Разделение труда и различие индивидов и их общественных функций вследствии сложной системы профессиональной специализации. И чем выше уровень разделения труда, тем сильнее у людей стремление к единству (к солидарности).

_**Мнение Эмиля Дюркгейма:**_ 

- Чем органичнее общество, тем выше его склонность к демократии, потому что последняя основана на свободе выбора, уважении к личности, защите прав человека. 
- И, напротив, чем более механическим является общество, тем более оно склоняется к тоталитаризму.

## Структурный функционализм Толкотта Парсонса (1902-1979 гг.)

---------
_**Вопросы по подкатегории:**_

7. Что представляет собой общество с точки зрения структурного функционализма Т. Парсонса?

---------

---------

7. _**Что представляет собой общество с точки зрения структурного функционализма Т. Парсонса?**_

---------

**Общество с точки зрения Толкотта Парсонса** - это социальная система, достигшая по отношению к окружающей среде наивысшего уровня самодостаточности, считая, что социальная эволюция является частью живых систем.

---------

_Параллель устанавливается между:_

- Процесс возникновения человека как биологического вида
- Процесс появления общества современного вида

Все виды, согласно мнению биологов принадлежат к одному виду, поэтому также можно считать, что все общества произошли от одного типа общества, которое эволюционируя проходило различные стадии.

---------

Стадии эволюции общества по Толкотту Парсонсу:

1. Примитивная
2. Продвинуто-примитивная
3. Промежуточная
4. Современная

---------

8. _**Какие функции, по мнению Т. Парсонса, обеспечивают выживание общества как системы?**_

---------

Принципиально важным в теории Толкотта Парсонса было утверждение о том, что все общества как социальные системы для выживания должны обладать набором из четырых основных функций:

- Адаптация
- Целедостижение
- Интеграция
- Удержание образца

---------

_**Функция - наиболее подходящая подсистема общества**_

**Адаптация - Экономика**

Чтобы выжить - любая социальная система должна приспосабливаться или адаптироваться как в внутренней ситуации, так и и к изменениям внешней среды. Данной функции соответсвует такая подсистема, как экономика, которая поставляет и распределяет материальные продукты, удовлетворяя определенные физиологические потребности всех членов общества.

**Целедостижение - Политика**

Система определения основных целей общества и поддержание процесса их достижения.

**Интеграция - Государственный аппарат**

Система связывания и увязывания всех социальных компонентов общества, обеспечения его устойчивости, внутреннего единства и солидарности.

**Удержание образца - Семья, школа, церковь**

Любая социальная система должна стремиться к сохранению и поддержанию своих основных социальных ценностей, являющихся основой для формирования социальных норм как образцов социального поведения индивидов.

---------

В своих трудах Толкотт Парсонс уделял значительное внимание проблеме социального порядка, полагая, что важнейшая социальная ценность общества - **консенсус (согласие)**, поскольку только оно может обеспечить социальное равновесие общества.

Основными условиями достижения социального равновесия, по мнению Толкотта Парсонса, являются:

1. Социализация, посредством которой общесвтенный ценности передаются от одного поколения к другому
2. Разнообразные механизмы социального контроля

## Сиволический интеракционизм Джорджа Герберта Мида (1863-1931 гг.)

---------

_**Основные этапы:**_

1. Простейшее бессознательное общение при помощи жестов, которое генетически развивается в ходе долгой эволюции.

2. Создания начинают постепенно осозновать действия друг друга на основании интерпретации жестов, за каждым из которых закрепляется определенное социальное значение.

_На этом этапе по мнению Джорджа Мида общий мир существует в той степени, в которой сущетсвует и **общий опыт**, позволяющий понимать значения жестов._

3. К символике добавляется язык, формирующий систему абстрактных символов.

_На этапе существования абстрактных символов (языка) становится возможным общение и понимание без предварительного общего опыта._

_**Вопросы по подкатегории:**_

9.  Какие идеи лежат в основе символического интеракционизма Д.Мида? 

---------

_**Идеи:**_

- Общество можно объяснить только путем рассмотрения принципов поведения людей, лежащих в основе их взимодействия
- Термин "символический" обозначает, что это социологическая школа делает акцент на "смысле", который вкладывают действующие в обществе лица ("актеры"), в тот или иной "символ", когда они вступают во взаимодействие "интеракцию".

**Социальное взаимодействие** - межличностная коммуникация ("обмен символами"), важнейшей особенностью которой признается способность человека "принимать роль другого", представлять, как его воспринимает партнер по общению и, соответственно, интерпретировать и конструировать свои собственные действия.

Из социального взаимодействия, следует, что развитие личности происходит в процессе взаимодействия с другими людьми в обществе, которое трактуется как **система непосредственных коммуникаций.**

_**Итог:**_ Таким образом, символический интеракционизм рассматривает общество с точки зрения поведения индивидов, вовлеченных в акты поведения, взаимодействия.

## Конфликтология

---------

_**Приверженцы данного направления (внесли вклад в данное направление):**_

Карл Макрс, Фридрих Энгельс, Макс Вебер, Людвиг Гумплович, Георг Зиммель

---------

_**Основная идея:**_

Видим в обществе не сбалансированность мотивов и взаимных интересов, а борьбу различных групп и идей, которая формирует существующие социальные структуры и отношения через конфликты.

_**Вопросы по подкатегории:**_

10.	Какова природа социального конфликта в концепциях П. Сорокина, К. Маркса,  Р. Дарендорфа? 

---------

_Существует 2 концепции конфликтологии:_

---------

- Марксистская
- Неомарксистская

---------

**Марксистская концепция (К.Макрс)**

---------

**У конфликтов есть предел, когда они могут закончиться.**

---------

Основное противоречие общества - является **собственность**, выступающая в качестве движущей силы социальных изменений, а основой социального конфликта является борьба за собственность.

_Если упростить:_

- **Объект (природа)** марксистской концепции - собственность
- **Предмет** марксистской концепции - борьба за собственность

---------

=> Ликвидация собственности снимает противоречия и создает условия для бесконфликтного развития общества 

=> Есть временное состояние общества и возможно достижение такого уровня общественного развития, когда социальные конфликты исчезнут.

---------

**Неомарксистская концепция**

---------

**Общества без конфликтов - не существует, так как конфликты являются неотъемлемой частью бытия и главным стимулом общественного развития**

---------

**Конфликт** - не аномалия, а норма, необходимый элемент социальной жизни, дающий выход социальной напряженности.

---------

_**Природа конфликта по мнению Питирима Сорокина:**_

---------

Источник конфликта лежит в подавлении базовых потребностей человека, без удовлетворения которых он не может существовать: в пище, одежде, жилье, самосохранении, творчестве, самовыражении и др.

_**Особенность:**_ Причина не в самих потребностях, а в средствах их удовлетворения, в их доступности.

---------

_**Природа конфликта по мнению Ральфа Дарендорфа (1929-2009):**_

---------

Источник конфликта лежит в политическом факторе, т.е. борьбе за власть, престиж, авторитет. Конфликт, по его мнению, рождается из того, что одна социальная группа сопротивляется "давлению" или господству противоположной ей социальной силы.

_**Задача общества:**_
Задача общества в том, чтобы направить конфликты в определенное русло, которое не разрушало бы всю систему, а вело бы к ее плановой эволюции. Для этого нужно сделать конфликты предметом открытых дискуссий, исходя из представления о том, что общество сохраняется как целое благодаря постоянному разрешению присущих ему внутренних конфликтов, являющихся оборотной стороной интеграции.

Главная задача: управление конфликтами, умение их контролировать.
