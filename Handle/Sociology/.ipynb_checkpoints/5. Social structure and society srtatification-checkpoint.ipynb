# Разбор

1. **Что такое социальная структура общества?**

-----------

**Социальная структура общества** - устойчивая связь определенных элементов. которые и формируют само общество как целостную социальную систему. При этому каждый из элементов социальный структуры, в свою очередь, явялется сложной системой со своими подсистемами и связями.

-----------

2. Какие элементы составляют социальную структуру общества?

-----------

Выделяют два социальный элемента в обществе, образующих структуру оного:

- Индивиды (первичный элемент)
- Социальные группы индивидов (вторичный элемент)

**Первичный элемент** - индвиды, которые занимают определенные позиции - статусы. При этом каждый индивид является частью общего, а именно - социальной группы.

**Вторичный элемент** - ссоциальные группы индивидов, образованные на основе общности социального признака. Данными признаками социальные группы не только объединяются, но и отличаются от других людей.

**П:** Пол, возраст, уровень дозодов, род деятельности, этническая или рассовая принадлежность.

**Социальный признак** - дифференцирующий критерий, который указывает на различное положение людей в обществе по отношению друг к другу и подчеркивает имеющееся в нем социальное неравенство, которое существует в обществах любых типов.

**П:** Даже в самых примитивных, где имущественные различия -> min существуют различия между мужчинами и женщинами.

3. Что такое страта?
4. Что такое социальная стратификация?
5. Каковы критерии социальной стратификации в концепциях М.Вебера, Р.Дарендорфа, Т.Парсонса?

-----------

**Страта** - тоже самое, что и "класс".

**Социальная стратификация** - совокупность социальных слоев общества. Она отражает естественную социальную неоднородность общества, неодинаковость социального положения его членов, их социальное неравенство, иерархический, устойчивый характер, постоянно воспроизводится. модифицируется и обеспечивает упорядоченное существование любого общества.

_Социальную стратификацию различные ученые разделялипо разным критериям (признакам):_

Были введены такие показатели стратификации, как

**М.Вебер:**

- Социальный престиж (складывается из образовательных, профессиональных, квалифцированных и иных признаков)
- Объем власти
- Размер собственности


**Р.Дарендорф:**

- "Авторитет"

2-ве **подгруппы** управляющих:

- Управлящие собственники
- Управляющие несобственники (бюрократы-менеджеры)

2-ве основные **подгруппы**:

- высшую - ("рабочая аристократия")
- низшую - низко квалифицированные рабочие, между которыми находится промежуточный "новый средний класс"

**П.Сорокин:**

Считал, что дать единую совокупность критериев принадлежности к какой-то страте - невозможно.

Выделял 3 стратифицированные парамиды:

1. Экономическая - деление на имущих и не имущих
2. Политическая  - наличие управляющих и управляемых
3. Профессиональную - разделение на раздичные группы по роду деятельности, занятием, обусловленное неравенством в престиже профессии.

Позиции могут не совпадать:

Глава государства, имеющий высший политический статус, в экономической стратификации может иметь более низкую позицию, по сравнению с миллиардером. И наоборот.

6. Что такое социальный статус?
7. Какие критерии лежат в основе иерархии статусов в социальной стратификации современного общества?

-----------

В конце 19 века английский историк Мейн термину "статус" социалогическое значение:

**Социльный статус** - это социальное положение, занимаемое человеком или социальной группой в обществе.

**Социльный статус** - это разносторонняя и, в то же время, обобщающая характеристика социального положения индивида, определяемая всем набором его социальных признаков, которые факсирует положение индивида в социальном пространстве, т.е. позицию индивида в обществе.

-----------

_Социальный статус определяется показателями:_

- Прирожденный (приписываемый) статус
- Достижимый (приобретенный) статус

**Прирожденный статус** - определяется социальными признаками, которые даются человеку от рождения.

**П:** Пол, национальные и расовые признаки, социальное происхождение (вытекающее из социального положения родителей).

_Почему биологические признаки вдруг относятся к социальным?_

Цвет кожи, пол, даются природой и являюстя биологическими характеристиками человека. Однако люди наделяют их социальным смыслом, делая основанием для социального неравнества.

**Достижимый статус** - те признаки, которые приобретаются (т.е. достигаются) человеком в течении жизни.

**П:** Професссия, образование, квалификация, род деятельности, уровень дозодов, должность и т.д.

**Социальная стратификация** - это совокупность социальных слоев (статусов), определенным образом ориентированных внутри социальной структуры.

Факторы, которые влияют на иерархию статусов в обществе, являются:

- уровень доходов
- объем власти
- уровень образования
- престижность профессии

-----------

8. Что такое социальная мобильность?
9. Что такое  горизонтальная  и вертикальная мобильность?
10. Что такое межпоколенная и внутрипоколенная мобильность?

-----------

**Социальная мобильность** - всякое изменение индивидом или социальной группой места в социальной структуре. Впервые ввел Питирим Сорокин.

-----------

**Горизонтальная социальная мобильность** - это переход индивида из одной социальной группы в другую, расположенную на одном и том же уровне, т.е. без перемены статуса.

**П:** смена работы в той же должности, смена места учебы

**Вертикальная социальная мобильность** - это перемещение индивида из одного социального слоя в другой, сопровождающееся изменением социального статуса.

**П:** Повышение в звании, в должности, в уровне дозодов и т.д.

-----------

**Межпоколенная (интергенерационная) мобильность** - определяется сравнением социального статуса родителей и из детей в определенный момент карьеры тех и других (по рангу их профессии, уровню образования и дозодов в приблизительно одинаковом возрасте)


**Внутрипоколенная (интрагенерационная) мобильность (социальная карьера)** - сравнение социального статуса одной личности в течение жизни.

-----------

11. Что такое социальная аномия?
12. Что такое  коллективная мобильность?

-----------

**Коллективная социальная мобильность** - это массовые изменения индивилами, социальными группами своего места в социальной структуре, осуществляемые как по горизонтали, так и по вертикали.
