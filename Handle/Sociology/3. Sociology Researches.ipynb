## Подготовка социологического исследования 

----------

1. Какова структура эмпирического социологического  исследования?

----------

**Эмпирическое исследование** - система логически последовательных методологических, методических и организационно-технических процедур, связанных между собой в целях получения достоверных данных об изучаемом социальном явлении или процессе.

_Эмпирическое (прикладное) социологическое исследование состоит из 3-ех основных этапов:_

----------

1. Подготовительный этап
2. Основной этап
3. Завершающий этап

----------

_В подробностях:_

1. **Подготовительный этап** - разработка программы прикладного исследования
2. **Основной этап** - сбор эмпирической информации
3. **Завершающий этап** - обработка и анализ полученных в ходе прикладного социологического исследования данных, а также формируются выводы и рекомендации.

Этап разработки программы конкретного социологического исследования является чрезвычайно важным, так как именно этот документ задает стратегию эмпирического исследования.

----------

1. _**Всякое научное исследование начинается с обоснования проблемы, в ходе которой формулируется исследовательский вопрос.**_

----------

**Проблема** - это всегда противоречие между знаниями и потребностями людей в решении каких-то задач и незнанием путей исредств их реализации вследствии отсутствия или недостаточности знания о механизме функционирования изучаемых предмета и объектов исследования.

_Что значит поставить проблему?_

**Поставить проблему** - значит выявить, обосновать и описать имеющее место противоречие.

**Решить проблему** - значит получить новое знание или создать теоретическую модель, объясняющую то или иное явление, выявить факторы, позволяющие воздействовать на развитие явления в желаемом направлении.

----------

3. _**Что означает проблема в социологическом исследовании?**_

----------

**Проблема в социологическом исследовании** - это существующая в самой реальности, в окружающей нас жизни противоречивая ситуация, носящая массовый характер и затрагивающая интересы больших социальных групп, либо социальных институтов, либо общества в целом, природа которой не вполне ясна.

Важной задачей на стадии разработки программы конкретного социологического исследования является перевод проблемной ситуации в постановку проблемы.

----------

_Для того, чтобы из проблемной ситуации сформулировать проблему, необходимо:_

----------

1. Установить реальное наличие данной проблемы:

----------

_Вопросы:_

1. Есть ли показатель, кол-венно или качественно характеризующий данную проблему?
2. Есть ли учет и статистика по этому показателю?
3. Достоверны ли учет и статистика по этому показателю?

----------

2. Вычленить наиболее существенные моменты, элементы или факторы проблемы, решение которых принадлежит социологии, а не другим наукам.

3. Определить какие социальные группы и личности связаны с этой проблемой?

4. Вычленить уже известные элементы проблемной ситуации, которые не требуют специального анализа и выступают как информационная база для рассмотрения неизвестных элементов.

5. Выделить в проблемной ситуации главные и второстепенные компоненты, чтобы определить основное направление поиска.

6. Проанализировать уже имеющиеся решения аналогичных проблем. С этой целью необходимо изучить всю литературу по данному вопросу и провести беседу с экспертами.

2. _**Определение цели и задач социологического исследования**_

----------

4. Как формулируется цель в социологическом исследовании?

----------

**Цель** - вытекает из общей направленности исследования. Всегда ориентированна на разрешения поставленной задачи.

**Задачи социологического исследования** - это совокупность конкретных целевых установок, направленных на разрешение проблемы. Они формулируются в соответсвии с целью исследования и определяют предполагаемые пути и средства разрешения исследуемой проблемы.

----------

3. _**Определение объекта и предмета исследования**_

----------

5. _**Что такое предмет и объект в социологическом исследовании?**_

----------

**Предмет исследования** - социальная сфера, внутри которой находится исследуемая проблема. Включает в себя те стороны и свойства объекта, которые в наиболее полном виде выражают исследуемую проблему и скрывающиеся в ней противоречия.

**Объект исследования** -люди, вовлеченные в исследуемый социальный процесс и являющиеся носителями той или иной социальной проблемы, т.е. носителями исследуемого социального противорчения.

----------

7. _**Что такое генеральная совокупность?**_
8. _**Что такое выборочная совокупность?**_

----------

1. Объект исследования не велик => сплошное исследование
2. Объект исследования многочисленен => процедура выборки необходимой совокупности.

**Генеральная совокупность** - совокупность всех возможных носителей исследуемой социальной проблемы.

**Выборочная совокупность (Выборка)** - часть объектов из генеральной совокупности, отобранная с помощью специальных приемов и правил.

_Вариации выборки:_

----------

- Вероятностная
- Случайная
- Гнездовая
- Целенаправленная
- Стихийная

----------

**Репрезентативность** - это свойство выборочной совокупности воспроизводить параметры и элементы генеральной совокупности.

**Ошибка репрезентативности** - расхождения между генеральной совокупностью и выборочной совокупностью.

Если выборка собрана правильно, то результаты исследования можно будет распространить (экстраполировать) на всю генеральную совокупность (весь объект исследовани).

Если выборка собрана НЕ правильно => ошибка будет преумнажаться в процессе исследования и значительно искажать полученные результаты.

Поэтому в программе социологического исследования должно быть четко указано:

----------

1. Каков объект эмпирического исследования
2. Является исследование сплошным или выборочным
3. Если оно является выборочным, то какова выборка, как она формировалась и претендует ли исследование на репрезентативность.

----------

4. Формулирование гипотез исследования

----------

**Гипотеза** - это научное предположение, сформулированное в утвердительной форме, выдвигаемое для объяснения каких-либо фактов, которые нужно в ходе исследования подтвредить или опровергнуть.

_По-другому:_ 

**Гипотеза** - это научное предположение о структуре социальных объектов, о характере и сущности связей между социальными явлениями, подлежащих исследованию.

----------

6. _**Какова роль гипотез в социологическом исследовании?**_

----------

Роль гипотезы - чрезвычайно велика. Гипотезы конкретизируют цель исследования и отражают его общую направленность.

Дальнейшие этапы эмпирического социологического исследование определяют - выдвинутые гипотезы, поскольку они обуславливают логику всего процесса исследования и последующего анализа результатов.

Чтобы справиться с данной задачей, гипотезы должны соответствовать следующим требованиям:

1. Иметь четкую и однозначную формулировку
2. Логически согласовываться с проблемой и целью
3. Заключать понятия, получившие предварительное уточнение и интерпретацию
4. Быть приложимыми к данным, заключенным в предварительном описании предмета исследования
5. Иметь возможность быть эмпирически провернеными.

----------

9. _**Что означает процедура интерпретации основных понятий?**_
10. _**Что означает процедура операционализации основного понятия и для чего она нужна?**_

----------

Важное место в программе исследования занимает логический анализ основных понятий, которые будет применяться в исследовании.

**Процедура интепретации основных понятий** - процедура заключается в объяснении содержания и стуктуры основных понятий, т.е. их интепретации, которая осуществляется с помощью расчленения (операционализации) основных понятий на составные элементы, способные описать их содержание.

_**Коротко:**_

**Процедура интепретации основных понятий** - осуществляет сбор кол-венной информации

**Операционализация** - о чем следует собирать информацию.

Задачи логического анализа:

1. Достижение однозначности в понимании содержания основного понятия.
2. Выявление совокупности операциональных понятий, на которые "раскладывается" основное понятие

----------

11. _**Что такое индикаторы в социологическом исследовании?**_

----------

**Понятия-индикаторы** - понятия, обозначающие элементарные фрагменты, которые могут быть исследованны кол-венными методами и доступны учету и регистрации.

Их определение помогает со следующими вопросами:

- Каким образом можно измерить необходимую социальную информацию?
- Как правильно сформулировать вопросы ответы на них в различных видах инструментария?
