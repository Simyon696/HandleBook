**Валовый внутренний продукт** - это размер всей экономики, сколько экономика в целом произвела в течение какого-то периода времени и сколько было продано в тот же год совершенно различным экономическим игрокам.

**Инфляция** это тоже велечина, единая для всей экономики, и на нее также влияют все те действия, которые производят совершенно различные экономические игроки в течении опять-таки какого-то периода времени.

Как мы увидим - в макроэкономике было время в течении нескольких десятилетий, когда у нее была своя уникальная методология (кейнсианская), отличающаяся от других , от других разделов экономики, но к сегодняшенму времени - методологические отличия - **стерлись**. Теперь эти отличия - чисто тематические.

В макроэкономике, как и в экономике в целом, существуют различные силы:

1. **Рыночные силы**, которые в значительной степени определяют то, что происходит с тем же ВВП или с той же самой инфляцией, или с обменными курсами. Данным силам довольно сложно восприпятствовать, и есть отдельные рассуждения, нужно или не нужно как-то препятствовать им.

2. **Государство проводит экономическую политику** - из этого рождается прикладное значение макроэкономики:

_Нужно понять как государству реагировать на события, происходящие в экономике._

Государство, реагируя на то, что происходит в экономике, также влияет на то, что происходит в экономике.

Так как на экономику влияет огромное кол-во факторов, таких как:

1. Рынки
2. Действие частных людей, коих огромное кол-во как в рамках одной страны, так и в мире
3. Действия экономических политиков, которые либо в ответ на действие частных людей, либо просто так, но тем не менее, принимают какие-то решения, которые дальше влияют на то, что происходит.

Нужно научиться это все отличать друг от друга:
_Что ялвяется результатом действия рыночных сил, а что результатом экономической политики_

_**Для кого данные знания нужны?**_

П: Все эти вещи важны для инвесторов, которые должны принимать решения:

1. Куда им направить свои деньги, а куда лучше не стоит
2. От чего зависят обменные курсы
3. От чего зависит инфляция
4. Возможность прогнозировать данные переменные.

_**Что для этого нужно?**_

Для того, чтобы прогнозировать эти переменные, нужно понимать их природу - их источник колебаний, источник движения как _обменных курсов_, так и _инфляции_, так и _прозводства_ в целом.

_Для того, чтобы углубиться в природу данных вопросов и научиться лучше прогнозировать велечины, макроэкономисты разделили данную науку на 2-ва очень больших подраздела:_

---

1. _**Долгосрочный экономический рост**_

Простой пример: Мы понимаем, что мы сегодня живем гораздо богаче, что это было несколько сотен лет назад. И очень важно понять природу этого роста, тем более, что этот рост неравномерен в мире:

- Есть страны, которые растут быстро, а есть те, которые растут медленно.
- Есть страны, которые стали богатыми, а есть те, которые остались бедными.

Понимание данных тенденций - тоже является важной частью экономики.

**Большая часть курса посвящена именно этой теме.**


2. _**Краткосрочный экономический рост (Краткосрочные колебания)**_

То есть мы видим, что экономический рост происходит неравномерно. В какие-то годы экономика падает, в какие-то растет аномально быстро.

Самый недавний период, который будет изучаться в рамках данного курса - **_мировой финансовый кризис_**, который начался в _**2008 году**_ и привёл к падению мирового ВВП. Некоторые страны, такие как Греция - пострадали очень сильно от этого кризиса, а некоторые другие гораздно меньше, но ВВП упал во всех странах. Но с точки зрения макроэкономики - это _**краткосрочное колебание**_.

Данное колебание довольно сильное, его важно уметь понять и объяснить, но тем не менее это не долгосрочный эффект.

Важное отличие долгосрочного экономического роста от краткосрочного состоит по мнению многих экономистов в том, что - _долгосрочный экономический рост определяется тем, что **экономика увеличивает свои производственные возможности.**_ То есть сегодня экономика умеет проихводить больше, чем она умела производить век назад.

Одни из возможных причин:

1. Новые технологии
2. Мы накопили производственные мощности в виде зданий и сооружений, и оборудования, и инфраструктуры, которых опять-таки не было 100 лет назад.

Именно увеличивающиеся возможности движут экономикой в долгосродчной перспективе.



_Различия развивающихся и развитых стран:_

1. **Для развивающихся стран** тема долгосрочного экономического роста наиболее важна в первую очередь , потому что развивающиеся страны пытаются стать богатыми, пытаюся догнать развитый мир.
Именно по этой причине вот такие вопросы краткосрочных колебаний или так называемого экономического цикла в развивающихся странах - они занимают меньшую часть общественной дискуссии.

2. **Для развитых стран** наибольшей заботой является плавный рост экономики, чтобы не было падений в тот или иной квартал или тот или иной год. У них меньше времени уделяется общественным дискуссиям долгосрочного экономического роста.

Но изучать мы будем обе темы в рамках данного курса, ибо в 2009 году экономика России резко упала на 8% и это было ощутимое падение, которое привело к временному росту безработицы и банкротсву многих предприятий. Поэтому эта тема совсем не является для нас второстепенной или неважной.
