### Понятие экономического цикла

Развитие экономики сопровождается конъюктурными волнами разной протяженности. Различают (выделяют) Э. циклы длиной:

- 7-11 лет (циклы Жугляра - промышленные циклы)
- 15-20 лет (Строительный цикл (здание служит примерно 20 лет) - Саймон Кузнец)
- 40-60 лет (Полувековые волны Николая Дмитриевич Кондратьева. Кондратьевские циклы)

Деловой (экономический) цикл - колебания уровня экономической активности (фактический реальный ВВП), когда периоды спада сменяются периодами подъема экономики.

_Экономисты выделяют 3 подхода к объяснению причин циклов:_

---------

- внешними (экзогенные, экстернальными) факторами (изобретение инноваций, войны, политичские факторы)
- внутренними (эндогенными, интернальными) факторами
- синтезом тех или иных факторов

---------
